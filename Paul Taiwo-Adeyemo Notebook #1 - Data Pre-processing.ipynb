### Introduction

Diabetes, a medical condition that arises from the inablity of the body to properly breakdown and store sugar causes 24.8 deaths per 100,000 population (as of 2020) in North America and costs  $237 billion yearly in direct medical costs in the US alone. Reducing this medical/human cost is an area in which much work is critically needed. This project aims at using machine learning to reduce that cost by emloying a three-tier preemptive approach.

#### Business Question

The aim of this analysis is to build machine learing models to mitigate the medical and human cost associaated with diabetes. This is carried out in a three tier approach listed below: 

> 1. Is a non-diabetic person at risk of diaabetes based on the person's lifestyle choices? 
> 2. Is a diabetic individual at risk of getting re-hospitalized?
> 3. Will the diabetic individual at risk of being re-hospitalized be hospitalized in less than 30 days or more than 30 days?

#### Value-add / Real Life Use

The models that will be produced from the project could potentially be used to reduce the medical and human costs of diabetes by:
> Predicting whether or not a person is at risk of diabetes and then recommend lifestyle changes that would put the individual out of the category of being at risk of developing diabetes.
>
> Predicting whether or not a diabetic individual would be re-hospitalized and then offer recommendations for medication changes that would put the diabetic individual out of the category of being at risk for diabetes
>
>For those in the risk group of getting hospitalized, predicting whether or not the individual would be readmitted in less than 30 days or more than 30 days. Then offer recommendations to put the individual out of the less than 30 days group and into the more than 30-day group.

```python
#function to define each categorical column
#sourced from lecture notes by BrainStation
def explore_categorical(col_name, df): 
    if df[col_name].dtypes == 'object':
        unique_cat = len(df[col_name].unique())
        print()
        print(f"\033[1m\033[94m{col_name} \n{20 * '-'}\033[0m") 
        print("Feature '{col_name}' has {unique_cat} unique categories".format(col_name=col_name, unique_cat=unique_cat))
        print()
        print(df[col_name].value_counts(normalize=True))
        sns.countplot(y=col_name, data=df, order = df[col_name].value_counts().index, color="#58508d")
        print()
```

```python
#make function to decode the ICP 9 codes
def map_diagn(column):
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(a) for a in range(1,140)], value ='Infection_&_Parasitic')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(b) for b in range(140,240)], value ='Neoplasms')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(c) for c in range(240,280)], value ='Endocrine_Immune')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(d) for d in range(280,290)], value ='Blood')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(e) for e in range(290,320)], value ='Mental_Disorders')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(f) for f in range(320,390)], value ='Nervous_System')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(g) for g in range(390,460)], value ='Circulatory_System')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(h) for h in range(460,520)], value ='Respiratory_System')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(i) for i in range(520,580)], value ='Digestive_System')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(j) for j in range(580,630)], value ='Urinigenital_System')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(k) for k in range(630,680)], value ='Pregnancy')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(l) for l in range(680,710)], value ='Skin')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(m) for m in range(710,740)], value ='Musculoskeletal_System')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(n) for n in range(740,760)], value ='Congenital_Anomalies')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(o) for o in range(760,780)], value ='Perinatal')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(p) for p in range(780,800)], value ='Ill-defined_Conditions')
    dataset1_df[column]=dataset1_df[column].replace(to_replace = [str(q) for q in range(800,1000)],value ='Injury_&_Poisoning')
    dataset1_df[column]=dataset1_df[column].str.replace('^V\d+' , 'Supplementary')
```

```python
#create function to visualize correlation
#Code sourced from BrainStation Lectures
def visualize_correlation(X):

    # calculate all the correlations between each of the independent variables
    corr_df = X.corr()

    # mask
    mask = np.triu(corr_df)


    # heatmap
    plt.figure(figsize=(10, 10))
    sns.heatmap(corr_df.round(2), vmax=1, vmin=-1, center=0, annot=True, mask=mask, cmap='coolwarm')
```

#### Feature Engineering

A new feature, COPD (Chronic Obstructive Pulmonary Disease)was made from Smoker, HeartDiseaseorAttack, HighBP and HighChol because these are features that are a part of COPD spectrum. Moreso, these features all have a high colinearity with each other as observed from the correlation chart below.

```python
#combine Smoker, HeartDiseaseorAttack, HighBP and HighChol to form COPD
dataset2_df['COPD'] = dataset2_df['Smoker'] + dataset2_df['HeartDiseaseorAttack'] +\
                      dataset2_df['HighBP'] + dataset2_df['HighChol']
```

```python
#make COPD features into a dataframe and check correlation
X = dataset2_df[['Stroke','HeartDiseaseorAttack', 'Diabetes_binary','Smoker', 'HighChol','COPD']]
```

```python
#visualize correlation
visualize_correlation(X)
```

Another new feature, Mental_General_Health was made by combining the GenHealth and the MentalHealth features, these features describe the a very similar health parameter and it makes sense to combine them. 

```python
#combine GenHealth and the MentalHealth features
dataset2_df['Mental_General_Health'] = dataset2_df['GenHlth'] + dataset2_df['MentHlth']
```

```python
#drop the redundant columns, that have been used in the new features
drop = ['HighChol','HeartDiseaseorAttack','Smoker','HighBP','MentHlth', 'GenHlth']
dataset2_df = dataset2_df.drop(drop, axis=1)
```
